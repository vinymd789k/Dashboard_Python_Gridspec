```python
#Importando Bibliotecas
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#Definindo os estilos do dashboard.
print(plt.style.available)
plt.style.use('seaborn')
```

```python
#Gerando os dados.
df = {"id" : np.arange(0, 20), "valor" : np.random.randn(20), "desconto" : np.random.randn(20), "local" : np.arange(20, 40)}
df = pd.DataFrame(df)
df
```

```python
#Definindo os espaço figure.
figure = plt.figure(figsize = (10, 6))
gs_master = gridspec.GridSpec(3, 2, height_ratios = [2, 2, 2])

#Primeira Grade.
gs_1 = gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec = gs_master[0, :])
title_axes = figure.add_subplot(gs_1[0])
title_axes.set_title("Valor dos Locais")
title_axes.bar(x = df["local"], height = df["valor"])
title_axes.set_xlabel("Local")
title_axes.set_ylabel("Valor")
title_axes.set_xticks(df["local"])
title_axes.set_xlim(left = 19)
figure.suptitle("Análise Valor-Desconto por ID-Local", fontsize = 20)

#Segunda Grande.
gs_2 = gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec = gs_master[1, :])
id_valor = figure.add_subplot(gs_2[0])
id_valor.bar(x = df["id"], height = df["valor"])
id_valor.set_title("Valor dos ID")
id_valor.set_xlabel("ID")
id_valor.set_ylabel("Valor")
id_valor.set_xticks(df["id"])
id_valor.set_xlim(left = -1)

# #Terceira Grade.
# Pimeiro Plot.
gs_31 = gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec = gs_master[2, :1])
local_valor = figure.add_subplot(gs_31[0, 0])
local_valor.scatter(x = df["local"], y = df["desconto"])
local_valor.set_title("Local X Desconto")
local_valor.set_xlabel("Local")
local_valor.set_ylabel("Desconto")
local_valor.set_xticks(df["local"])
local_valor.set_xlim(left = 19)

# Segundo Plot.
gs_32 = gridspec.GridSpecFromSubplotSpec(1, 1, subplot_spec = gs_master[2, 1])
local_desconto = figure.add_subplot(gs_32[0, 0])
local_desconto.scatter(x = df["id"], y = df["desconto"])
local_desconto.set_title("Id x Desconto")
local_desconto.set_xlabel("Id")
local_desconto.set_ylabel("Desconto")
local_desconto.set_xticks(df["id"])
local_desconto.set_xlim(left = -1)

# Harmonizando o layout.
gs_master.tight_layout(figure)
plt.show()
```
